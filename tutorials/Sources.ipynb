We would like to be able to check the kcsd on different sources.

These sources should be:
- realistic
- simple and/or complex
- iterative
- pseudorandom
- with different noise
- able to check majority of shapes in a systematic way
- not to big

Sources are to be a 'true csd' assumption, that we can check how the reconstruction works. 

One will calculate potentials, using a forward model, and then reconstruct the csd and compare it with the 'true csd'

We assume that:
- the recording space is from 0 to 1 at maximum - it would scale to real life. 


### a simple source could look like this:

```python
import sources
from matplotlib import pyplot as plt
import numpy as np

#libsize, depending on how many sources we want to check, can be 'small', 'medium' and 'large'

def plot1dSource(csd):
    resolution = 100
    axis = np.linspace(-0.5, 1.5, resolution*2) # resolution is between 0 and 1, and we want our csd to be from -0.5 to 1.5 
    plt.plot(axis, csd)
    plt.axvline(1, lw=0.2, color='k')
    plt.axvline(0, lw=0.2, color='k')
    ymax = 1.1
    if np.min(csd)< -.1:
        ymin = -1.1
    else:
        ymin = -0.1
    plt.ylim(ymin, ymax)

Counter = 1# goes from 0
test_source = sources.SingleGauss(Counter, dim=1, libsize='small', resolution=100)

plot1dSource(test_source)

plt.show()
```

```python

# one can use this to generate a library of sources:

Counter = 0
plt.figure(figsize=(20,20))
for i in range(9):
    test_source = sources.SingleGauss(Counter, dim=1, libsize='small', resolution=100)
    if test_source is False:
        continue
    Counter+=1
    plt.subplot(3,3,i+1)#, aspect='equal')
    plot1dSource(test_source)
        

plt.show()
```

```python
# Dipol

Counter = 0
subplot_counter = 1
plt.figure(figsize=(20,20))
for i in range(10):
    test_source = sources.dipol(Counter, dim=1, same_sign=True, libsize='small', resolution=100)
    if test_source is False:
        continue
    Counter+=1
    plt.subplot(5,5,subplot_counter)#, aspect='equal')
    subplot_counter+=1
    plot1dSource(test_source)

for i in range(10):
    test_source = sources.dipol(Counter, dim=1, same_sign=False, libsize='small', resolution=100)
    if test_source is False:
        continue
    Counter+=1
    plt.subplot(5,5,subplot_counter)#, aspect='equal')
    subplot_counter+=1
    plot1dSource(test_source)

plt.show()
```

```python
# Tripole 1d

Counter = 0
subplot_counter = 1
plt.figure(figsize=(20,20))
for i in range(10):
    test_source = sources.tripole(Counter, sign_way=1, libsize='small', resolution=100)
    if test_source is False:
        continue
    Counter+=1
    plt.subplot(5,5,subplot_counter)#, aspect='equal')
    subplot_counter+=1
    plot1dSource(test_source)
   # plt.axes().set_aspect('equal')

for i in range(10):
    test_source = sources.tripole(Counter, sign_way=2, libsize='small', resolution=100)
    if test_source is False:
        continue
    Counter+=1
    plt.subplot(5,5,subplot_counter)#, aspect='equal')
    subplot_counter+=1
    plot1dSource(test_source)
    
plt.show()
```

```python
# Random shape sources
# those are pseudorandom sources, with a given seed - source of certain number will always be the same.
# They are to be simmilar to different, complex shapes that are hard to describe.
# A pool big enough can check if there exist a case, where a method is not working

Counter = 0
subplot_counter = 1
plt.figure(figsize=(20,20))
for i in range(25):
    test_source = sources.RandomCSD1d(Counter, resolution=100)
    if test_source is False:
        continue
    Counter+=1
    plt.subplot(5,5,subplot_counter)#, aspect='equal')
    subplot_counter+=1
    plot1dSource(test_source)


plt.show()
```

```python
# the aim is to be able to iterate for a whole library of sources.
# Here is a function, that called will return the functions above
Counter = 30

test_source = sources.Source1D(Counter, libsize = 'small')

plot1dSource(test_source)
plt.show()
```
